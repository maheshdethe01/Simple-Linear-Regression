```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import r2_score, mean_squared_error 
```

```python
data = pd.read_csv("NewspaperData.csv") # reading the dataset
data.head()
```

```python
data.info()
```

# Correlation

```python
data[['daily','sunday']].corr('pearson') # correlation 
```

```python
sns.hisplot(data['daily'])
```

```python
sns.histplot(data['sunday'])
```

Fitting a Linear Regression Model

```python
sns.regplot(x="daily", y="sunday", data=data);
```

```python
# labelling the dataset
X = data.daily.values.reshape(-1,1)
y = data.sunday.values.reshape(-1,1)
```

```python
# splitting the data 
X_train, X_test, y_train, y_test = train_test_split(X,y,train_size=0.9,random_state=42)
```

```python
# fitting the data into the simple linear regression model
lr = LinearRegression()
lr.fit(X_train,y_train)
```

```python
print(lr.coef_)
print(lr.intercept_)
```

```python
y_pred = lr.predict(X_test)
```

```python
print(f'MSE: {mean_squared_error(y_pred,y_test)}')
print(f'r2_Score: {r2_score(y_pred,y_test)}')
```

```python
# As the newspaper linear model has 0.87 r2 score , we can use this model for predicting sunday newspaper sales.
```
